```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format='retina'
plt.style.use('ggplot')
```

```python
success = 2
failure = 98
sim0 = beta.rvs(success + 1, failure + 1, size=10000)
sim1 = beta.rvs(success + 2, failure + 1, size=10000)
sim2 = beta.rvs(success + 1, failure + 2, size=10000)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
sns.kdeplot(sim0, ax=ax, clip=[0, 1], label='Default, Beta(2, 98)')
sns.kdeplot(sim1, ax=ax, clip=[0, 1], label='+1 Success')
sns.kdeplot(sim2, ax=ax, clip=[0, 1], label='+1 Failure')
ax.legend()
# ax.set(xscale='log')
fig.tight_layout()
```
